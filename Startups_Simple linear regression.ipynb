```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '50_Startups.csv'  # File name of your dataset
df = pd.read_csv('/content/50_Startups.csv')

# Group by 'State' and calculate the mean for the specified columns
grouped_data = df.groupby('State')[['Profit', 'Marketing Spend', 'Administration', 'R&D Spend']].mean()

# Reset index for plotting
grouped_data = grouped_data.reset_index()

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Number of bars
states = grouped_data['State']
x = range(len(states))  # X positions for states

# Bar width
bar_width = 0.2

# Plot each variable
ax.bar(x, grouped_data['Profit'], width=bar_width, label='Profit', align='center')
ax.bar([i + bar_width for i in x], grouped_data['Marketing Spend'], width=bar_width, label='Marketing Spend', align='center')
ax.bar([i + bar_width * 2 for i in x], grouped_data['Administration'], width=bar_width, label='Administration', align='center')
ax.bar([i + bar_width * 3 for i in x], grouped_data['R&D Spend'], width=bar_width, label='R&D Spend', align='center')

# Set the x-axis with state names
ax.set_xticks([i + bar_width * 1.5 for i in x])
ax.set_xticklabels(states)

# Labels and Title
ax.set_xlabel('State', fontsize=14)
ax.set_ylabel('Mean Values', fontsize=14)
ax.set_title('Comparison of Profit, Marketing Spend, Administration, and R&D Spend by State', fontsize=16)
ax.legend()

plt.tight_layout()
plt.show()
```

```
# prompt: remove outliers for all columns

import pandas as pd
import numpy as np

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Assuming 'df' is your DataFrame (replace with your actual DataFrame name)
# Example usage: Remove outliers from 'Profit'
df = remove_outliers_iqr(df, 'Profit')

# Example usage: Remove outliers from all numerical columns in one go

numerical_cols = ['Profit', 'Marketing Spend', 'Administration', 'R&D Spend']
for col in numerical_cols:
    df = remove_outliers_iqr(df, col)
```
